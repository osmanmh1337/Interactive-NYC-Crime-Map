By splitting the Bronx into smaller subdivisions we can flesh out the economic disparities and differences in arrest rates within the borough. We will use data directly from the US Census Bureau and the NYPD. To aggregate this data into a bar graph, we can perform a series of transformations and merges where we now have a dataframe with a column for median income, and another column for the # of arrests corresponding to that median income. To make it easier to visualize we will put these data points into larger intervals of 10,000 (ex: \$0-9,999, \$10,000-19,999, etc.). We can now make a bar graph using the matplotlib library.



We can calculate and plot the correlation between median income and arrest rates
by resampling the arrests to yearly so it matches the yearly median income data points. We will also reformat the dataframe to be in date time formate and merge them. Afterwards we can simply plot the correlation over time using matplotlib and identify and trends/periodicity.

```python
#DATA WRANGLING AND MANIPULATION
def extract_numeric_part(column_name): # Remove all characters except digits and '.' to make the data uniform
    numeric_part = ''.join(filter(lambda x: x.isdigit() or x == '.', column_name))
    return numeric_part
def check_and_append_decimal(column_name): # Append decimals if needed to make the data uniform
    if '.' not in column_name:
        return column_name + '.00'
    else:
        return column_name


def format_column_names(column): #make it so the census tracts ID's are all 6 digits, which will match the census tract ID's in the shapefiles
    max_length = len(column)

    if max_length == 1:
        return '00000' + column
    elif max_length == 2:
        return '0000' + column
    elif max_length == 3:
        return '000' + column
    elif max_length == 4:
        return '00' + column
    elif max_length == 5:
        return '0' + column
    else:
        return column



#Manipulating the data so it can be merged with gpd, also getting rid of NaN and unwanted values (data cleaning)
df= pd.read_csv('/content/bronx-median-income.csv')
columns_to_keep = [col for col in df.columns if "Median income (dollars)!!Estimate" in col]
df = df[columns_to_keep]
df = df.loc[[1]].copy()
df.columns = df.columns.map(extract_numeric_part)
df.columns = df.columns.map(check_and_append_decimal)
mapping = {col: col.replace('.', '') for col in df.columns}
df.rename(columns=mapping, inplace=True)
df.columns = df.columns.map(format_column_names)
df = df.replace('-', '0')
df = df.replace('NaN',0)
row_index = 0
pattern = r'^\d+,\d+$'  # Pattern to match "any number of digits','any number of digits"

# Function to check and replace values based on the pattern, this will get rid of unwanted values that do not follow the desired format
def check_and_replace(row):
    updated_row = []
    for value in row:
        if isinstance(value, str) and re.match(pattern, value.strip()):
            updated_row.append(value)
        else:
            updated_row.append(0)
    return updated_row

row_values = df.iloc[row_index].tolist()
updated_values = check_and_replace(row_values)
df.iloc[row_index] = updated_values

df = df.rename(index={1: "median_income"})
df=df.T
df['median_income'] = df['median_income'].str.replace(',', '').astype(float)

#Merging median income df with bronx gdf, we can now create a chloropleth map which colors the different census tracts a deeper/lighter color depending on its median income value
df
```

```python
merged_gdf = bronx_gdf.merge(df, how='left', left_on='TRACTCE20', right_index=True)

merged_gdf
```

## ADDING CHLOROPLETH

```python
map_nyc = folium.Map(location=[40.8448, -73.8648], zoom_start=15)

# Adding choropleth, for coloring different census tracts based on median income value
folium.Choropleth(
    geo_data=merged_gdf,
    name='Median Income',
    data=merged_gdf,
    columns=['TRACTCE20', 'median_income'],
    key_on='feature.properties.TRACTCE20',
    fill_color='Purples',
    fill_opacity=0.9,
    line_opacity=0.5,
    legend_name='Median Income ($)',
    highlight=True,
    nan_fill_color='grey',
    bins=6,
    reset=True
).add_to(map_nyc)

map_nyc
```

```python
arrest_coords_original= pd.read_csv('/content/2020_arrest_coords.csv')
arrest_coords_original
```

```python
#Getting arrest coordinates so we can plot them on the map into their respective census tracts
arrest_coords = arrest_coords_original.copy()
arrest_coords = arrest_coords[arrest_coords['ARREST_BORO'] == 'B']

geometry = [Point(xy) for xy in zip(arrest_coords['Longitude'], arrest_coords['Latitude'])]
arrest_gdf = gpd.GeoDataFrame(arrest_coords, geometry=geometry, crs='EPSG:4326')

# Calculate centroids of census tracts in 'tracts_gdf', this is where we can place the label for # of arrests for each census tract
bronx_gdf['centroid'] = bronx_gdf['geometry'].centroid

joined_gdf = gpd.sjoin(arrest_gdf, bronx_gdf, how='inner', op='within')

# Group by TRACTCE20 and calculate number of arrests per census tract, so now we have one # for each census tract which represents the total number of arrest that fall in that tract
arrests_per_tract = joined_gdf.groupby('TRACTCE20').agg(
    Arrests_Count=('index_right', 'size'),  # Count number of arrests per tract
    Centroid=('centroid', 'first')
).reset_index()

arrests_per_tract
```

## ADDING LABELS FOR ARREST COUNT

```python
#Adding labels for count of arrest onto our map (map_nyc)
for index, row in arrests_per_tract.iterrows():
    centroid = row['Centroid']
    arrests_count = row['Arrests_Count']


    folium.Marker(
        location=[centroid.y, centroid.x],
        icon=folium.DivIcon(
            icon_size=(150, 36),
            icon_anchor=(7, 20),
            html=f'<div style="font-size: 10pt; color: black; font-weight: bold;">{arrests_count} </div>'
        )
    ).add_to(map_nyc)

map_nyc
```

```python
#Creating a bar graph that shows us the # of arrests for each income range, the ranges increment by 10,000 and start from 0 and go all the way to 140,000
df_grouped = merged_gdf.copy()
df1=arrests_per_tract.copy()
df= df_grouped.join(df1['Arrests_Count'], how='outer')
income_intervals = [
    (0, 9999),
    (10000, 19999),
    (20000, 29999),
    (30000, 39999),
    (40000, 49999),
    (50000, 59999),
    (60000, 69999),
    (70000, 79999),
    (80000, 89999),
    (90000, 99999),
    (100000, 109999),
    (110000, 119999),
    (120000, 129999),
    (130000, 139999)
]
def get_interval_label(income_range):
    return f"${income_range[0]}-${income_range[1]}"
df['income_interval'] = pd.cut(df['median_income'], bins=pd.IntervalIndex.from_tuples(income_intervals))
arrest_counts_by_income = df.groupby('income_interval')['Arrests_Count'].sum()
plt.figure(figsize=(10, 6))
arrest_counts_by_income.plot(kind='bar', color='skyblue')
plt.title('Number of Arrests by Median Income in Bronx, NY (2020)', fontsize=14)
plt.xlabel('Median Income($)', fontsize=12)
plt.ylabel('Number of Arrests', fontsize=12)
plt.xticks(ticks=range(len(income_intervals)), labels=[get_interval_label(interval) for interval in income_intervals], rotation=90)
plt.show()
```

## PLOTTING THE CORRELATION (median income vs arrest count)

```python
#Plotting the correlation between median income and # of arrests
median_income_yearly = median_income_df.resample('Y').last()
# Setting 'arrests_2013_2023' to yearly so it matches median_income_yearly in terms of period
arrests_yearly = arrests_2013_2023.resample('Y').sum()
#Merging median_income_yearly and _arrest_yearly so we can plot it
merged_df = pd.merge(median_income_yearly, arrests_yearly, left_index=True, right_index=True)
# Calculate correlation between 'median_income' and 'Arrest_Count'
correlation = merged_df['median_income'].corr(merged_df['Arrest_Count'])
print(f"Correlation between median income and arrest count: {correlation}")
plt.figure(figsize=(10, 6))
plt.plot(merged_df.index, merged_df['median_income'], label='Median Income')
plt.plot(merged_df.index, merged_df['Arrest_Count'], label='Arrest Count')
plt.xlabel('Year')
plt.ylabel('# of Arrests')
plt.title('Median Income vs. # of Arrests in Bronx, NY (2014-2023)')
plt.legend()
plt.show()
```

```python
#Plotting monthly arrests counts from 2013-2020 to try and identify any seasonal trends
monthly_arrests = arrests_2013_2023.resample('M').sum()
filtered_monthly_arrests = monthly_arrests[monthly_arrests.index < '2020-01-01']
plt.figure(figsize=(24, 6))
plt.plot(filtered_monthly_arrests.index, filtered_monthly_arrests['Arrest_Count'],linestyle='--',color='purple')
plt.xlabel('Year')
plt.ylabel('# of Arrests')
plt.title('Monthly Arrest Count in Bronx, NY (2013-2020)')
plt.axvline(pd.Timestamp('2013-01-01'), color='black', linestyle='-', linewidth=1.25) #adding lines to make it easier to see where the year starts and ends
plt.axvline(pd.Timestamp('2014-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2015-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2016-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2017-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2018-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2019-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.axvline(pd.Timestamp('2020-01-01'), color='black', linestyle='-', linewidth=1.25)
plt.show()
```
